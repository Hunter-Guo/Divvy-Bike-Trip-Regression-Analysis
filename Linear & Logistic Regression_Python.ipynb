# Step 3: Do Feature Engineering

```python
bounding_box = {'Downtown': (-87.651393, -87.596007, 41.866745, 41.905919),
                'Navy Pier': (-87.610168, -87.599876, 41.891737, 41.892654),
                'Lakefront': (-87.6300, -87.6000, 41.8500, 41.9500),
                'Hyde Park': (-87.6067, -87.5800, 41.7800, 41.8100)}
for prefix in ['start', 'end']:
    df[f'area_{prefix}'] = pd.Series(data='Other', index=df.index)
    for area, box in bounding_box.items():
        df.loc[df[f'longitude_{prefix}'].between(*box[:2]) & df[f'latitude_{prefix}'].between(*box[2:]), f'area_{prefix}'] = area

df['trip_minutes'] = df['tripduration']
```

# Step 6: Define Business Problem
## Business Problem:
> The Chicago Department of Transportation (CDOT) is interested in understanding how Chicago Divvy bike trip duration is related to other factors of the user, the trip, and the weather in Chicago.
## Task
> Build a linear regression model using historical Chicago taxi trip data from year 2022 with `trip_seconds` as the response variable and other useful information as explanatory variables.
## Some Considerations
- **Low R-Square**
  - If adding intercept, the model has very low R^2=0.2127, therefore we exclude the intercept
  - Excluding the intercept might not be a sensible thing to do here, and it makes sense that `temperature` is not highly correlated with `trip_minutes` intuitively
- **Log Transformation**
  - The distribution of `temperature` is roughly uniform and `temperature` has negative value, log transformation is not applicable
- **Feature Engineering**
  - `avg_dpcapacity` is derived from `dpcapacity_start` and `dpcapacity_end`
- **Others**
  - `temperature` and `avg_dpcapacity` are the only two numerical explanatory variables. They are not highly collinear with other predictors in the model since their VIF are both close to 1
  - A logistic regression model may be needed
