# 3 Simple Linear Regression

```r
model <- lm(trip_minutes ~ temperature, data = data)

summary(model)
```

```r
coef(model) # See str(model)
```

```r
summary(model)$r.squared # See str(summary(model))
```

```r
ggplot(data, aes(temperature, trip_minutes)) +
  geom_point(alpha = 0.1) +
  geom_smooth(method = 'lm',
              formula = y ~ x,
              se = FALSE,
              color='orange') +
  ggtitle('trip_minutes ~ temperature')
```

```r
# Compute mean squared error MSE
y <- data$trip_minutes
y.pred <- fitted(model) # Also fitted.values(model) and model$fitted.values
mean((y - y.pred) ** 2)
```

```r
# Compute coefficient of determination R^2
ss.reg <- sum((y.pred - mean(y)) ** 2)
ss.error <- sum((y - y.pred) ** 2)
ss.total <- sum((y - mean(y)) ** 2) # SST = SSR + SSE
              
c(ss.reg, ss.error, ss.total)
```

```r
c(cor(y, y.pred) ** 2,
  ss.reg / ss.total, # SSR / SST
  1 - ss.error / ss.total, # 1 - SSE / SST
  summary(model)$r.squared)
```

# 10 Logistic Regression

```r
# # Create the binary target variable
# data <- data %>%
#   mutate(trip_long = as.factor(ifelse(trip_minutes > 10, 1, 0)))
```

```r
factor(data$trip_minutes > 10)
```

```r
model <- glm(factor(trip_minutes > 10) ~ temperature * factor(year) + factor(month) * factor(day) +
           factor(area_start) * factor(area_end) + factor(usertype) + factor(gender) + factor(events) + avg_dpcapacity, family = binomial, data = data) 
# Logit link is default. See ?family
```

```r
final.model <- step(model, direction = 'backward', k = 2, test = 'Chisq')
```

```r
# Generate predicted probabilities
data$predicted <- predict(final.model, type = "response")

# Group data by avg_dpcapacity (or a chosen variable) for visualization
binned_data <- data %>%
  group_by(avg_dpcapacity = cut(avg_dpcapacity, breaks = 10)) %>%
  summarize(mean_trip = mean(trip_minutes > 10),
            avg_x = mean(avg_dpcapacity))

# Create the plot
ggplot(data, aes(x = avg_dpcapacity, y = as.numeric(trip_minutes > 10))) +
  geom_jitter(alpha = 0.1, color = "blue") +
  geom_smooth(aes(y = predicted), method = "glm", method.args = list(family = "binomial"),
              se = FALSE, color = "orange", size = 1) +
  geom_point(data = binned_data, aes(x = avg_x, y = mean_trip),
             color = "gray", size = 2, alpha = 0.8) +
  labs(title = "Logistic Regression",
       x = "Average DPCapacity",
       y = "Probability (Trip > 10 Minutes)") +
  theme_minimal()
```

```r
tau <- 0.5
y.hat <- as.numeric(fitted(final.model) > tau) # Convert TRUE/FALSE to 0/1

(confusion.matrix <- table(factor(data$trip_minutes > 10), y.hat))

n <- sum(confusion.matrix) # Total number of data points
tn <- confusion.matrix[1,1] # True negative
fp <- confusion.matrix[1,2] # False positive
fn <- confusion.matrix[2,1] # False negative
tp <- confusion.matrix[2,2] # True positive

(accuracy <- (tp + tn) / n) # Accuracy = 1 - error (or misclassification) rate
(tpr <- tp / (tp + fn)) # True positive rate = recall
(fpr <- fp / (fp + tn)) # False positive rate
(precision <- tp / (tp + fp)) # Precision
```

```r
library(ggplot2)

# Convert confusion matrix to a data frame for visualization
confusion_df <- as.data.frame(as.table(confusion.matrix))
colnames(confusion_df) <- c("Actual", "Predicted", "Count")

# Plot heatmap
ggplot(confusion_df, aes(x = Predicted, y = Actual, fill = Count)) +
  geom_tile(color = "black") +
  geom_text(aes(label = Count), color = "white", size = 5) +
  scale_fill_gradient(low = "white", high = "blue") +
  theme_minimal() +
  labs(title = "Confusion Matrix Heatmap",
       x = "Predicted",
       y = "Actual")
```

```r
# Create a data frame for the metrics
metrics <- data.frame(
  Metric = c("Accuracy", "True Positive Rate", "False Positive Rate", "Precision"),
  Value = c(accuracy, tpr, fpr, precision)
)

# Plot bar chart
ggplot(metrics, aes(x = Metric, y = Value, fill = Metric)) +
  geom_bar(stat = "identity", color = "black") +
  scale_fill_brewer(palette = "Set3") +
  theme_minimal() +
  labs(title = "Model Performance Metrics",
       x = "Metric",
       y = "Value") +
  geom_text(aes(label = round(Value, 2)), vjust = -0.5)
```
